```python
def insert_mirrored_rows(df, num_rows=30):
    """
    Insert chronologically mirrored data point at head and tail of df
    """
    df = df.copy()
    mirrored_rows_head = df.iloc[:num_rows].copy()
    mirrored_rows_head = mirrored_rows_head.iloc[::-1].reset_index(drop=True)

    mirrored_rows_tail = df.iloc[-num_rows:].copy()
    mirrored_rows_tail = mirrored_rows_tail.iloc[::-1].reset_index(drop=True)

    df_extended = pd.concat([mirrored_rows_head, df, mirrored_rows_tail], ignore_index=True)
    
    return df_extended
```

```python
def moving_avg(df, length=24):
    """ 
    Finds moving average for past 24 hour
    """
    df = insert_mirrored_rows(df.copy())
    # columns_to_edit = [col for col in df.columns if col != 'Date Time']
    # out_df = df.copy()
    # for col in columns_to_edit:
    #     out_df[col] = df[col].rolling(window=length).mean()
    out_df = df.set_index('Date Time').rolling(window=length).mean().reset_index()
    return out_df.iloc[30:-30].reset_index(drop=True)
    
def moving_avg_mid(df, lengths=[3]):
    """ 
    Finds centered moving averages for given window lengths.
    Overwrites the same columns on each iteration.
    """
    df = insert_mirrored_rows(df.copy())
    columns_to_edit = [col for col in df.columns if col != 'Date Time']
    numeric_df = df[columns_to_edit]
    for length in lengths:
        df[columns_to_edit] = df[columns_to_edit].rolling(window=length, center=True).mean()

    return df.iloc[30:-30].reset_index(drop=True)
def moving_avg_mid2(df, lengths=[3]):
    """ 
    Finds centered moving averages for given window lengths.
    Overwrites the same columns on each iteration.
    """
    df = insert_mirrored_rows(df.copy())
    for length in lengths:
        df = df.rolling(window=length, center=True).mean()

    return df.iloc[30:-30].reset_index(drop=True)
```

```python
def get_event(flow_df, precip_df):
    precip_df = precip_df.copy()
    flow_df = flow_df.copy()
    
    precip_df = moving_avg_mid(precip_df, [24])
    flow_df = moving_avg_mid(flow_df, [24])
    
    precip_df.set_index("Date Time", inplace=True)
    flow_df.set_index("Date Time", inplace=True)

    flow_col_name = flow_df.columns[0]
    precip_col_name = precip_df.columns[0]
    
    events = []
    in_event = False
    original_flow = flow_df[flow_col_name].min() * 18
    event_start = None
    
    for date in precip_df.index:
        if not in_event and precip_df.loc[date, precip_col_name] > 0.0:
            event_start = date
            in_event = True  
            
        if in_event:
            flow_value = flow_df.loc[date, flow_col_name] 
            
            if pd.isna(flow_value) or (flow_value <= original_flow * 1.5):
                event_end = date
                events.append((event_start, event_end))
                in_event = False  
    
    merged_events = []
    for i in range(len(events)):
        if i == 0:
            merged_events.append(events[i])
        else:
            prev_start, prev_end = merged_events[-1]
            curr_start, curr_end = events[i]
    
            if (curr_start - prev_end) <= timedelta(hours=6):
                merged_events[-1] = (prev_start, curr_end)
            else:
                merged_events.append(events[i])

    filtered_events = []
    for start, end in merged_events:
        if (end - start) >= timedelta(hours=18):
            max_flow = flow_df.loc[start:end, flow_col_name].max()
            if max_flow > 0.7:
                filtered_events.append((start, end))

    return filtered_events
```

```python
flow_data = MMSD_sewerflow_predicted_precipitation_df[['Date Time', 'MS0311 Flow']]
precip_data = weighted_precip_for_streamgage_df[['Date Time','40869416']]

flow_col_name = flow_data.columns[1]
precip_col_name = precip_data.columns[1]

events = get_event(flow_data, precip_data)

fig, ax = plt.subplots(figsize=(400, 5))

plt.plot(flow_data['Date Time'], flow_data[flow_col_name], linestyle='-', color='g', markersize=6, label="Data Line")
plt.plot(flow_data['Date Time'], moving_avg_mid(flow_data, [24])[flow_col_name], linestyle='-', color='y', markersize=6, label="Data Line")

flow_max = flow_data[flow_col_name].max()
flip_precip = (flow_max*1.4) - precip_data[precip_col_name]*4
plt.plot(precip_data['Date Time'], flip_precip, linestyle='-', color='r', markersize=6, label="Data Line")

# ax.plot(weighted_precip_for_streamgage_df.index, weighted_precip_for_streamgage_df["40869416"] * 3, label="Sewer Flow", color="blue")
# ax.plot(MMSD_sewerflow_predicted_precipitation_df.index, MMSD_sewerflow_predicted_precipitation_df["MS0311 Flow"], label="Precipitation (scaled)", color="green", alpha=0.6)

for start, end in events:
    ax.axvspan(start, end, color="gray", alpha=0.3)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.set_title("Sewer Flow and Precipitation Events with Merged Events")
ax.legend()
plt.grid(True)

plt.show()
```

```python

def top_10_events(flow_data, precip_data):
    events = get_event(flow_data, precip_data)
    df = flow_data
    df['40869416'] = precip_data['40869416']
    
    # Ensure the 'date' column is in datetime format
    
    # Calculate max value in each range
    range_max_values = []
    for start, end in events:
        mask = (df["Date Time"] >= start) & (df["Date Time"] <= end)
        max_val = df.loc[mask, "MS0311 Flow"].max()
        range_max_values.append(((start, end), max_val))
    
    # Sort by max value descending and get top 10
    top_10_ranges_only = [r[0] for r in sorted(range_max_values, key=lambda x: x[1], reverse=True)[:10]]
    return top_10_ranges_only
```

```python
def trim(df, event):
    return df[(df['Date Time'] >= event[0]) & (df['Date Time'] <= event[1] - pd.Timedelta(hours=1))].reset_index(drop = True)
    
```
