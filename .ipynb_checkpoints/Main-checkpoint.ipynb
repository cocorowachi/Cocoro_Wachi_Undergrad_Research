```python
def insert_mirrored_rows(df, num_rows=30):
    """
    Insert chronologically mirrored data point at head and tail of df
    """
    df = df.copy()
    mirrored_rows_head = df.iloc[:num_rows].copy()
    mirrored_rows_head = mirrored_rows_head.iloc[::-1].reset_index(drop=True)

    mirrored_rows_tail = df.iloc[-num_rows:].copy()
    mirrored_rows_tail = mirrored_rows_tail.iloc[::-1].reset_index(drop=True)

    df_extended = pd.concat([mirrored_rows_head, df, mirrored_rows_tail], ignore_index=True)
    
    return df_extended

def moving_avg_mid(df, lengths=[3]):
    """ 
    Finds centered moving averages for given window lengths.
    Overwrites the same columns on each iteration.
    """
    df = insert_mirrored_rows(df.copy())
    for length in lengths:
        df = df.rolling(window=length, center=True).mean()

    return df.iloc[30:-30].reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
from scipy.signal import find_peaks

def show_graph(events, flow_df, precip_df):
    i = 0
    for event in events:
        event = (event[0] - pd.Timedelta(hours=40), event[1] - pd.Timedelta(hours=1))
        flow_event_df = src.trim(flow_df, event).iloc[:, 1]
        precip_event_df = src.trim(precip_df, event).iloc[:, 1]
        l3 = src.luedke_3_level(flow_event_df, precip_event_df, 60, b1, b2, b3, b4, b5)
        bp1, bp2, bp3, bp4, bp5 = average_param[i]
        l3_fit = src.luedke_3_level(flow_event_df, precip_event_df, 60, bp1, bp2, bp3, bp4, bp5)
        hourly_range = pd.date_range(start=event[0], end=event[1] - pd.Timedelta(hours=1), freq='h')
        date_time_df = pd.DataFrame({'Date Time': hourly_range})

        flow_event_df.index = date_time_df['Date Time'].values
        precip_event_df.index = date_time_df['Date Time'].values
        l3.index = date_time_df['Date Time'].values
        l3_fit.index = date_time_df['Date Time'].values
        
        plt.figure(figsize=(15, 6))
        plt.plot(date_time_df['Date Time'], l3, color='b', markersize=4, label="Simulated flow")
        # plt.plot(date_time_df['Date Time'], l3_fit, color='r', markersize=4, label="best_fit")
        rate_of_change = flow_event_df.diff()*2
        
        plt.plot(date_time_df['Date Time'], flow_event_df, color='g', markersize=4, label="Observed flow")
        # plt.plot(date_time_df['Date Time'], rate_of_change, color='b', markersize=4, label="observed")


        
        flow_max = max(flow_event_df.max(),l3.max())
        flip_precip = (flow_max*1.4) - precip_event_df*1.5
        # plt.plot(date_time_df['Date Time'], flip_precip, linestyle='-', color='r', markersize=6, label="Precipitation")
        plt.xlabel("Date")
        plt.ylabel("Sewer Flow [MGD]")
        plt.title('Luedke 3 level model')
        plt.grid(True)
        plt.legend()

        r, p_value = pearsonr(flow_event_df, l3) # 4
        r_fit, p_value = pearsonr(flow_event_df, l3_fit) # 4
    
        stdev_sim = l3.std()
        stdev_sim_fit = l3_fit.std()
        stdev_obs = flow_event_df.std()

        C_b = (2*stdev_obs*stdev_sim)/((stdev_obs**2)+(stdev_sim**2)+(stdev_obs-stdev_sim)**2) # 5
        C_b_fit = (2*stdev_obs*stdev_sim_fit)/((stdev_obs**2)+(stdev_sim_fit**2)+(stdev_obs-stdev_sim_fit)**2) # 5
            
        print(f"Average Cinci: {r*C_b}")
        print(f"Fitted Cinci: {r_fit*C_b_fit}")

        
        total_precip = precip_event_df.sum()
        total_flow = flow_event_df.sum()
        avg_precip = total_precip/precip_event_df.count()
        avg_flow = total_flow/precip_event_df.count()*2.5
        print(f"avg precip: {avg_precip}")
        print(f"avg flow: {avg_flow}")
        
        plt.axhline(y=avg_flow, color='r', linestyle='--', label='Threshold')
        
        high_flow_times = flow_event_df[flow_event_df > avg_flow].index
        event_windows = [(t - pd.Timedelta(hours=8), t + pd.Timedelta(hours=24)) for t in high_flow_times]
        # Step 1: Create a boolean mask for times with flow > avg
        mask = flow_event_df > avg_flow
        high_flow_times = flow_event_df.index[mask]
        
        # Step 2: Merge consecutive high-flow times into continuous intervals
        event_windows = []
        if not high_flow_times.empty:
            start = high_flow_times[0]
            prev = high_flow_times[0]
        
            for current in high_flow_times[1:]:
                # Check if current time is exactly 1 hour after the previous
                if current - prev != pd.Timedelta(hours=1):
                    # Not consecutive: close the current window
                    event_windows.append((start, prev + pd.Timedelta(hours=1)))
                    start = current
                prev = current
            # Add the last window
            event_windows.append((start, prev + pd.Timedelta(hours=1)))
        
        # Step 3: Shade merged event periods
        for start, end in event_windows:
            plt.axvspan(start, end, color='gray', alpha=0.2)
        
        # Step 4: Filter data for only the event periods
        mask = pd.Series(False, index=flow_event_df.index)
        for start, end in event_windows:
            mask |= flow_event_df.index.to_series().between(start, end)
        
        filtered_flow_df = flow_event_df[mask]
        filtered_precip_df = precip_event_df[mask]
        filtered_avg_l3_df = l3[mask]
        filtered_fit_l3_df = l3_fit[mask]

        # plt.plot(filtered_flow_df.index, filtered_flow_df, color='b', markersize=8, label="best_fit")
        
        fil_r, p_value = pearsonr(filtered_flow_df, filtered_avg_l3_df) # 4
        fil_r_fit, p_value = pearsonr(filtered_flow_df, filtered_fit_l3_df) # 4
    
        fil_stdev_sim = filtered_avg_l3_df.std()
        fil_stdev_sim_fit = filtered_fit_l3_df.std()
        fil_stdev_obs = filtered_flow_df.std()

        fil_C_b = (2*fil_stdev_obs*fil_stdev_sim)/((fil_stdev_obs**2)+(fil_stdev_sim**2)+(fil_stdev_obs-fil_stdev_sim)**2) # 5
        fil_C_b_fit = (2*fil_stdev_obs*fil_stdev_sim_fit)/((fil_stdev_obs**2)+(fil_stdev_sim_fit**2)+(fil_stdev_obs-fil_stdev_sim_fit)**2) # 5
            
        print(f"Flow Average Cinci: {fil_r*fil_C_b}")
        print(f"Flow Average R: {fil_r}")
        print(f"Flow Fitted Cinci: {fil_r_fit*fil_C_b_fit}")
        print(f"Flow Fitted R: {fil_r_fit}")
        plt.show()

        plt.figure(figsize=(15, 6))
        plt.scatter(filtered_flow_df, filtered_avg_l3_df, color='b', label="avg_fit")
        sns.regplot(x=filtered_flow_df, y=filtered_avg_l3_df, ci=None, line_kws={"color": "green"}, label="Line of best fit")
        plt.plot(filtered_flow_df, filtered_flow_df, color='r', markersize=4, label="1:1 line")
        plt.title("Simulated vs. Observed data for conveyance interval")
        plt.xlabel("Observed Data [MGD]")
        plt.ylabel("Simulated Data[MGD]")
        plt.legend()
        plt.grid(True)
        plt.show()
        
        i+=1
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import seaborn as sns

# Example inputs
event_ranges = precip_events_top10_MS0311_df
param_values = average_param

# Set up plot
fig, ax = plt.subplots(figsize=(12, 6))

# Assign consistent colors for each parameter
colors = ['red', 'green', 'blue', 'orange', 'purple']

# Plot each event
for i, ((start, end), params) in enumerate(zip(event_ranges, param_values)):
    time_range = pd.date_range(start=start.replace(year=2000), end=end.replace(year=2000), freq='H')
    
    for j, param in enumerate(params):
        y_vals = [param] * len(time_range)
        ax.plot(time_range, y_vals, label=f'Param {j+1}' if i == 0 else "", color=colors[j], linewidth=2, alpha=0.8)

# Formatting
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
plt.xticks(rotation=45)
ax.set_xlabel('Time')
ax.set_ylabel('Parameter Value')
ax.set_title('Parameter Values Across Events')
ax.grid(True)
ax.legend(title='Parameters', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Generate sample data
index = pd.date_range('2025-01-01', periods=100, freq='H')
series = pd.Series(np.sin(np.linspace(0, 10*np.pi, 100)) + np.random.normal(0, 0.1, 100), index=index)

# Find peaks above a threshold
peaks, properties = find_peaks(series, height=0.8)
peak_times = series.index[peaks]

# Create 1-day-before to 1-day-after windows
event_windows = [(peak - pd.Timedelta(hours=2), peak + pd.Timedelta(hours=2)) for peak in peak_times]

# Plot
plt.figure(figsize=(12, 5))
plt.plot(series, label='Time Series')
plt.plot(peak_times, series.iloc[peaks], 'ro', label='Peaks > 0.8')

# Highlight windows
for start, end in event_windows:
    plt.axvspan(start, end, color='gray', alpha=0.2)

# Add a horizontal line at the threshold
plt.axhline(y=0.8, color='gray', linestyle='--', label='Threshold')

plt.title('Local Maxima with Event Windows Highlighted')
plt.legend()
plt.tight_layout()
plt.show()
```
