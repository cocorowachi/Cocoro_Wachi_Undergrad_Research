```python
for df in df_list:
    df['Date Time'] = pd.to_datetime(df['Date Time'])
```

```python
precip_df = pd.read_csv("Precipitation gauge coordinate.csv")
precipitation_stations = {}
for row in precip_df.itertuples():
    precipitation_stations[row.name] = (row.longitude, row.latitude)


stream_df = pd.read_csv("Stream gauge coordinate.csv")
sewerage_stations = {}
for row in stream_df.itertuples():
    sewerage_stations[row.name] = (row.longitude, row.latitude)
```

```python
def euclidean_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)
    
# Convert dictionary values (coordinates) into NumPy arrays for calculations
precipitation_points = np.array(list(precipitation_stations.values()))
sewerage_points = np.array(list(sewerage_stations.values()))

def weighted_average(precip_dict, sewer_dict, n_nearest=3):
    weighted_averages = {}

    for sewer_name, sewer in sewer_dict.items():
        distances = np.array([
            euclidean_distance(sewer[0], sewer[1], precip[0], precip[1]) 
            for precip in precip_dict.values()
        ])
        
        # Prevent division by zero by adding a small threshold
        distances = np.where(distances == 0, 1e-6, distances) 
        
        # Get the indices of the 3 nearest precipitation points
        nearest_indices = np.argsort(distances)[:n_nearest]
        
        # Compute weights (inverse of distance)
        weights = 1 / distances[nearest_indices]
        
        # Normalize weights so they sum to 1
        weights /= np.sum(weights)
        
        # Create weight dictionary with precipitation station names
        weight_dict = {name: 0 for name in precip_dict.keys()}  
        for idx, weight in zip(nearest_indices, weights):
            precip_name = list(precip_dict.keys())[idx]
            weight_dict[precip_name] = weight 
        
        # Store weights for the sewerage station
        weighted_averages[sewer_name] = weight_dict

    return weighted_averages
```

```python
def calculate(precipitation_stations, sewerage_stations):
    weights_dict = weighted_average(precipitation_stations, sewerage_stations)

    df_weights = pd.DataFrame(weights_dict)
    
    
    plt.figure(figsize=(8, 6))
    plt.scatter(precipitation_points[:, 0], precipitation_points[:, 1], c='blue', label='Precipitation Points')
    plt.scatter(sewerage_points[:, 0], sewerage_points[:, 1], c='red', label='Sewerage Points')
    
    for i, sewer in enumerate(sewerage_points):
        for j in np.argsort([euclidean_distance(sewer[0], sewer[1], precip[0], precip[1]) for precip in precipitation_points])[:3]:
            plt.plot([sewer[0], precipitation_points[j, 0]], [sewer[1], precipitation_points[j, 1]], 'k--', alpha=0.5)
    
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.legend()
    plt.grid(True)
    plt.show()
    # print(df_weights)
    # print(df_weights.T)
    return df_weights
```

```python
df_weights = calculate(precipitation_stations, sewerage_stations)
```

```python
weighted_precip_dict = {'Date Time': MMSD_precip_all_df['Date Time']}
weights_T_df = df_weights.T
cols = [col for col in MMSD_precip_all_df.columns if col != 'Date Time']
precip_num_df = MMSD_precip_all_df[cols]
precip_num_df.columns = precip_num_df.columns.str.split(' ').str[0]

for stream_col_name, weights in weights_T_df.items():
    weighted_avg = (precip_num_df * weights).sum(axis=1) / sum(weights)
    # print(stream_col_name, weighted_avg)
    # print()
    weighted_precip_dict[stream_col_name] = weighted_avg
weighted_precip_df = pd.DataFrame(weighted_precip_dict)
print(weighted_precip_df.head(5))

# csv = weighted_precip_df.to_csv(f'MMSD_weighted_precip_for_streamgage_df.csv', index = False) 
```
